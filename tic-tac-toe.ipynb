```python
def sum(x,y,z):
    return x+y+z

def tie(xin):
    for i in xin:
        if xin[i]==0:
            return -1
    print(" Game TIE***************** START ANOTHER***")
    return 1


def game(xin,yin):
    zero='X' if xin[0] else('O' if yin[0] else 0)
    one='X' if xin[1] else('O' if yin[1] else 1)
    two='X' if xin[2] else('O' if yin[2] else 2)
    three='X' if xin[3] else('O' if yin[3] else 3)
    four='X' if xin[4] else('O' if yin[4] else 4)
    five='X' if xin[5] else('O' if yin[5] else 5)
    six='X' if xin[6] else('O' if yin[6] else 6)
    seven='X' if xin[7] else('O' if yin[7] else 7)
    eight='X' if xin[8] else('O' if yin[8] else 8)
    
    print(f"{zero} | {one} | {two} ")
    print(f"--|---|--")
    print(f"{three} | {four} | {five} ")
    print(f"--|---|--")
    print(f"{six} | {seven} | {eight}")
    print('***********************************')



def winr(xin,yin):
    wins=[[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8],[0,4,8],[2,4,6]]
    for win in wins:
        if (sum(xin[win[0]],xin[win[1]],xin[win[2]])==3):
            print("X win the Game")
            return 0
        if (sum(yin[win[0]],yin[win[1]],yin[win[2]])==3):
            print("Y win the Game")
            return 1
        
    return -1
            


if __name__=='__main__':
    turn=1
    xin=[0,0,0,0,0,0,0,0,0]
    yin=[0,0,0,0,0,0,0,0,0]
    print("WELCOME! to tic-tac-toe")
    game(xin,yin)
    while True:
        if turn==1:
            x=int(input("X enter your choice"))
            if x>8:
                print(" Enter valid choice")
                continue
            if (xin[x]==1) or (yin[x]==1):
                print(" Already filled choose another")
                continue
            xin[x]=1
            game(xin,yin)
            turn=0
        elif turn==0:
            y=int(input("Y enter your choice"))
            if y>8:
                print(" Enter valid choice")
                continue
            if (xin[y]==1) or (yin[y]==1):
                print(" Already filled choose another")
                continue
            yin[y]=1
            game(xin,yin)
            turn=1
        k=winr(xin,yin)
        w=tie(xin)   
        if (k!=-1) or (w==1) :
            break


    
    
```
